#### About this dataset: https://data.novascotia.ca/Communications/News-Releases-Issued-by-Government-of-Nova-Scotia/xcif-vvr3

```python
df = pd.read_csv("https://raw.githubusercontent.com/tarjin/TextMining/master/BrownHound.csv")
```

```python
df = df.dropna(subset=['Post'])
```

```python
df.head()
```

```python
df.tail()
```

```python
text = df['Post']
```

```python
type(text)
```

```python
text = text.to_string()
```

```python
wordcloud = WordCloud(scale=2, width = 200, height=200, collocations=True, colormap='Greens', max_words=200, background_color="orange",max_font_size=40, relative_scaling=.5).generate(text)
```

```python
plt.rcParams['figure.figsize']=(20,200)
```

```python
plt.figure()
```

```python
plt.imshow(wordcloud)
```
