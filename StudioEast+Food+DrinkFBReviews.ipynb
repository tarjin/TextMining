```python
from wordcloud import WordCloud
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/tarjin/TextMining/master/StudioEastFBReviews.csv")
```

```python
df = df.dropna(subset=['Post'])
```

```python
df.head()
```

```python
df.tail()
```

```python
text = df['Post']
```

```python
type(text)
```

```python
text = text.to_string()
```

```python
wordcloud = WordCloud(scale=2, width = 200, height=200, collocations=True, colormap='Greys', max_words=100, background_color="white",max_font_size=40, relative_scaling=.5).generate(text)
```

```python
plt.rcParams['figure.figsize']=(20,200)
```

```python
plt.figure()
```

```python
plt.imshow(wordcloud)
```
